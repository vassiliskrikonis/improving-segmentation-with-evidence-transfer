# Skyline12 Dataset

```python
from datasets.skyline12 import Skyline12

skyline12 = Skyline12('datasets/skyline12/data')
```

Without augmentation:

```python
from itertools import islice

for x, y, z in islice(skyline12, 5):
    Skyline12.show_sample(x, [y, z])
```

With preprocessing (classes and scribbles for individual buildings are squashed into a single class)  and augmentation:

```python
ds = skyline12.as_tf_dataset(folds=2, keep_individual=False)

for x, y, z in islice(ds, 5):
    Skyline12.show_sample(x, [y, z], from_tensors=True)
```

Each fold is a pass through all the samples of the original datasets with random augmentations so that no two samples are identical.

```python
orig_len = len(skyline12)
ds = skyline12.as_tf_dataset(folds=2)
for i, (x, _, _) in enumerate(ds):
    if i % orig_len == 0:
        Skyline12.show_sample(x, from_tensors=True)
```

For reproducibility, each instantiation of the dataset and each tf.data.Dataset, produces the same results, depending on the given seed.

```python
ds1 = skyline12.as_tf_dataset(folds=1)
x, _, _ = next(iter(ds1))
Skyline12.show_sample(x, from_tensors=True)
```

```python
ds2 = skyline12.as_tf_dataset(folds=1)
x, _, _ = next(iter(ds2))
Skyline12.show_sample(x, from_tensors=True)
```

```python
skyline12_1 = Skyline12('datasets/skyline12/data', random_state=0)
ds = skyline12_1.as_tf_dataset(folds=1)
x, _, _ = next(iter(ds))
Skyline12.show_sample(x, from_tensors=True)
```

```python
skyline12_2 = Skyline12('datasets/skyline12/data', random_state=0)
ds = skyline12_2.as_tf_dataset(folds=1)
x, _, _ = next(iter(ds))
Skyline12.show_sample(x, from_tensors=True)
```
